2. Pull total number of orders that were completed on 18th March 2023 with the first name John and last name Doe

```python
SELECT SUM(Quantity)
FROM SALES
WHERE Date = '2023-03-18'

SELECT *
FROM CUSTOMERS
WHERE First_name LIKE 'John' AND Last_name LIKE 'DOE'
GROUP BY Date;
```
